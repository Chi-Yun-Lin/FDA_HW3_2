## 進行分檔

- 由 `df.info()` 可以知道共有 5456 筆 entries，所以分為:
    - index 0 to 4909: `train.csv`，共 4910 筆
    - index 4910 to 5455: `test.csv`，共 546 筆，但會缺少資料欄位名稱，所以會額外再新增欄位名稱在第一列，總共 547 筆

```python
import os
from os import path

# 確認檔案是否已存在
if path.exists('train.csv') == True:
    os.remove('train.csv')
if path.exists('test.csv') == True:
    os.remove('test.csv')

cond = 0
cut = 4910

# 將資料切成兩部分: train data : test data = 9 : 1
with open('google_review_ratings.csv', 'r', encoding='UTF-8') as file:
    for line in file:
        with open('train.csv', 'a', encoding='UTF-8') as file_1:
            with open('test.csv', 'a', encoding='UTF-8') as file_2:
                if cond < cut:
                    if cond == 0:
                        file_2.write(line + '\n')       # 在 test.csv 加入欄位名稱
                    file_1.write(line + '\n')
                else:
                    file_2.write(line + '\n')
        cond += 1
```
